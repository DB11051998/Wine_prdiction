#### treatment of outliers 

```python
df_train.describe()
```

```python
print(df_train['chlorides'].quantile(0.10))
print(df_train['chlorides'].quantile(0.80))
```

```python
df=pd.DataFrame(columns=['new'])
```

```python
df['new']=np.where(df_train["chlorides"] <0.031,0.031,df_train['chlorides'])
df['new']=np.where(df_train["chlorides"] >0.073, 0.073,df_train['chlorides'])
```

```python
df_train['chlorides']=df['new']
```

```python
df_train['chlorides'].skew()
```

```python
for i in df_train.columns[0:12]:
    if df_train[i].skew()>1:
        df[i]=np.where(df_train[i] <df_train[i].quantile(0.20),df_train[i].quantile(0.20),df_train[i])
        df[i]=np.where(df_train[i] >df_train[i].quantile(0.90),df_train[i].quantile(0.90),df_train[i])
```

```python
df.skew()
```

```python
for i in df.columns:
        df_train[i]=df[i]
```

```python
df_train.skew()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4,figsize=(15,15))
for i, column in enumerate(df_train.columns[0:12]):
    sns.distplot(df_train[column],ax=axes[i%3,i%4])
```

```python
X=pd.concat([df_train[df_train.type.eq('RED')], df_white[df_train.type.eq('WHITE')].iloc[:2000]], axis=0).sample(frac=1).reset_index(drop=True)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X.drop(columns=['type']),X.type,test_size=0.33,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
pred=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```

## It's a Nice accuracy score 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi=LogisticRegression()
```

```python
logi.fit(X_train,y_train)
```

```python
pred=logi.predict(X_test)
```

```python
accuracy_score(y_test,pred)
```
